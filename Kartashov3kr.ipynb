```python
#Третє завдання. Варіант 2.
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline

# Задані точки
i = np.array([0, 1, 2, 3, 4])
x = np.array([0.4, 0.6, 0.9, 1.4, 2])
y = np.array([2.45, 1.63, 0.95, 0.73, 1.95])

# Побудова кубічного сплайну
cs = CubicSpline(x, y)

# Генерація значень для гладкого графіка
smooth_x = np.linspace(min(x), max(x), 100)
smooth_y = cs(smooth_x)
```

```python
#Графік
plt.scatter(x, y, label='Табличні дані')
plt.plot(smooth_x, smooth_y, label='Кубічний сплайн')
plt.title('Кубічний сплайн для заданої функції')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
