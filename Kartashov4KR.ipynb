```python
#Четверте завдання. Варіант 2
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import least_squares

def f(x):
    return x * np.sin(x)

x_data = np.arange(0.1, 1.1, 0.1)
y_data = f(x_data)

def linear_approximation(x, params):
    return params[0] * x + params[1]

def quadratic_approximation(x, params):
    return params[0] * x**2 + params[1] * x + params[2]

initial_params_linear = [1, 1]
initial_params_quadratic = [1, 1, 1]

result_linear = least_squares(lambda params: y_data - linear_approximation(x_data, params), initial_params_linear)
result_quadratic = least_squares(lambda params: y_data - quadratic_approximation(x_data, params), initial_params_quadratic)

y_linear = linear_approximation(x_data, result_linear.x)
y_quadratic = quadratic_approximation(x_data, result_quadratic.x)
```

```python
#Графік
plt.plot(x_data, y_data, 'o', label='f(x)')
plt.plot(x_data, y_linear, label='Лінійне наближення')
plt.plot(x_data, y_quadratic, label='Квадратичне наближення')
plt.title('Наближення методом найменших квадратів')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
