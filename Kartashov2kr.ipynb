```python
#Друге завдання. Варіант 2.
import numpy as np
import matplotlib.pyplot as plt

def lagrange_interpolation(x_values, y_values, x):
    result = 0
    for i in range(len(y_values)):
        term = y_values[i]
        for j in range(len(x_values)):
            if j != i and (x_values[i] - x_values[j]) != 0:
                term = term * (x - x_values[j]) / (x_values[i] - x_values[j])
        result += term
    return result

x_values = [-3, 2, 0, 2, -2.5, -1, 1.5, 3]
y_values = [-13, 3, 5, 7, 2, 1, 3, 5]

interpolated_values = [lagrange_interpolation(x_values, y_values, x) for x in x_values]
smooth_x_values = np.linspace(min(x_values), max(x_values), 100)
smooth_interpolated_values = [lagrange_interpolation(x_values, y_values, x) for x in smooth_x_values]

plt.scatter(x_values, y_values, color='red', label='Задані точки')
plt.plot(smooth_x_values, smooth_interpolated_values, label='Інтерполяційний багаточлен')
plt.title('Гра')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
